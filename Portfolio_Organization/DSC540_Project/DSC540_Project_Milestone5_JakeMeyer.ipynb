### Milestone 5: Merging the Data and Storing in a Database/Visualizing Data

```python
'''
Import the necessary libraries to complete Milestone 5.
''' 
import numpy as np

import pandas as pd

import sqlite3

import matplotlib.pyplot as plt

import sqlalchemy
```

### Read the Final DataFrames from Milestones 2, 3, and 4

```python
'''
The final DataFrames from Milestones 2,3, and 4 were stored as CSV files.
Move a copy of the CSV files into the working directory.
Read the files with read_csv() and store as pandas DataFrame.
'''

milestone2_df = pd.read_csv("df_final_milestone2.csv")
milestone3_df = pd.read_csv("Milestone3_final_df.csv")
milestone4_df = pd.read_csv("Milestone4_df_final.csv")
```

```python
'''
Create a 'city_state' relationship column to join the DataFrame on for milestone2_df.
'''
milestone2_df['city_state'] = milestone2_df['city'] + '-' + milestone2_df['state']
```

```python
'''
View first 5 rows of milestone2_df with head().
'''
milestone2_df.head()
```

```python
'''
Add 'state' column back into milestone3 to assist with the merge of the tables in SQLite.
Moved the 2021_state_ranks.csv file back into the working directory for this file.
Need to add the column 'state' based on abbreviation for milestone3_df.
Create a dictionary of the state abbreviations based on 'state_name'.
'''
state_ranks = pd.read_csv("2021_state_ranks.csv")
state_dict = dict(zip(state_ranks['state_name'],state_ranks['state']))
```

```python
'''
Use map() to create the column 'state' within state_df. Use the state_dict recently created as the map argument.
Removed 'city' from this data set as it was not an essential/valuable column.
'''
milestone3_df['state'] = milestone3_df['state_name'].map(state_dict)
```

```python
'''
View first 5 rows of milestone3_df with head().
'''
milestone3_df.head()
```

```python
'''
Create a 'city_state' relationship column to join the DataFrame on for milestone4_df.
'''
milestone4_df['city_state'] = milestone4_df['city'] + '-' + milestone4_df['state']
```

```python
'''
View first 5 rows of milestone4_df with head().
'''
milestone4_df.head()
```

### Load the DataFrames into SQL Database as 3 Tables

```python
'''
Create an engine with sqlalchemy.create_engine() for milestone5.db.
'''
engine = sqlalchemy.create_engine('sqlite:///milestone5.db')
```

```python
'''
Create 'milestone2_table' in sqlite database milestone5.db.
Use df.to_sql() to perform the action.
'''
milestone2_df.to_sql('milestone2_table',engine, if_exists='replace', index=False)
```

```python
'''
Use pd.read_sql() to show the table contents in milstone5.db.
'''
pd.read_sql('SELECT * FROM milestone2_table', engine)
```

```python
'''
Create 'milestone3_table' in sqlite database milestone5.db.
Use df.to_sql() to perform the action.
'''
milestone3_df.to_sql('milestone3_table',engine, if_exists='replace', index=False)
```

```python
'''
Use pd.read_sql() to show the table contents in milstone5.db.
'''
pd.read_sql('SELECT * FROM milestone3_table', engine)
```

```python
'''
Create 'milestone4_table' in sqlite database milestone5.db.
Use df.to_sql() to perform the action.
'''
milestone4_df.to_sql('milestone4_table',engine, if_exists='replace', index=False)
```

```python
'''
Use pd.read_sql() to show the table contents in milstone5.db.
'''
pd.read_sql('SELECT * FROM milestone4_table', engine)
```

```python
'''
Merge the three tables in sqlite by 'city-state' for milestone2_table and milestone4_table and by 'state'
for milestone3_table. Store the merged tables as milestone5_df.
'''
milestone5_df = pd.read_sql('SELECT * FROM milestone2_table,milestone3_table,milestone4_table '
                        'WHERE milestone2_table.city_state = milestone4_table.city_state '
                        'AND milestone3_table.state = milestone4_table.state', engine)
```

```python
'''
View the first 5 rows of milestone5_df with .head().
'''
milestone5_df.head()
```

```python
'''
Review the columns of milestone5_df to ensure all tables were merged together.
'''
milestone5_df.columns
```

```python
'''
Drop the repeated columns in milestone5_df.
Ensure the first column is retained.
'''
milestone5_df = milestone5_df.loc[:,~milestone5_df.T.duplicated(keep='first')]
```

```python
'''
Understand the shape of milestone5.
'''
milestone5_df.shape
```

### Export Milestone5_df to a CSV File

```python
'''
Export milestone5_df to a csv file.
Label the file as Milestone3_final_df.csv
'''
milestone5_df.to_csv('Milestone5_df.csv',index=False)
```

```python
'''
This section of the code will utilize a for loop to plot
all the histograms for the essential columns using the plt.hist() function. 
'''
df_reduced = milestone5_df[['2021_us_news_overall_score',
       '2021_quality_of_life', '2021_value',
       '2019_us_news_overall_score', '2019_data_scientist_job_postings', 
       'average_salary_state', 'hourly_rate_state', 'job_count_state',
       'average_salary_city', 'hourly_rate_city',
       'population', 'population2010', 'race_white', 'race_black',
       'race_ameri_es', 'race_asian', 'race_hawaiian_pi', 'race_hispanic',
       'race_other', 'race_multi', 'males', 'females', 'age_under_5',
       'age_5to9', 'age_10to14', 'age_15to19', 'age_20to24', 'age_25to34',
       'age_35to44', 'age_45to54', 'age_55to64', 'age_65to74', 'age_75to84',
       'age_over_85', 'median_age', 'household_count',
       'average_household_size', 'families_count', 'average_family_size',
       'house_units_count', 'vacant_house_count', 'owner_occupied_house',
       'renter_occupied_house', 'population_growth']]

# For loop generation of histogram plot across all columns listed as important above.
for col in df_reduced.columns:
    # Iterate across columns in df_reduced to plot a histogram of each col.
    plt.hist(df_reduced[col]) 
    plt.grid(True)
    plt.title("Histogram of {}".format(col), fontsize=15)
    plt.xlabel("Variable " +col, fontsize=10)
    plt.xticks(fontsize=10) 
    plt.yticks(fontsize=10) 
    plt.show() 
```
