### Background
My final project will focus on Major League Baseball (MLB). The question I intend to analyze is, “Do MLB players with high salaries tend to perform better than MLB players with lower salaries?” The salary scale varies substantially throughout the league with some players making over 20 million dollars and others making under 200 thousand dollars. Organizations must invest in players based on their skillset and potential to get better. I would like to understand if the players paid the most also perform the best. Some additional research questions to support this hypothesis are discussed below:

* Does position play a role in salary? (i.e. do pitchers make more than fiel position players)
* Do both offensive and defensive performance influence salary?
* Are there players that are at the top of the MLB pay scale that play below average?
* Are there players at the lower end of the MLB pay scale that play above average?

```python
def MakeNormalModel(salary):
    """Plots a CDF with a Normal model.

    salary: sequence
    """
    cdf = thinkstats2.Cdf(salary, label='salary')

    mean, var = thinkstats2.TrimmedMeanVar(salary)
    std = np.sqrt(var)
    print('n, mean, std', len(salary), mean, std)

    xmin = mean - 4 * std
    xmax = mean + 4 * std

    xs, ps = thinkstats2.RenderNormalCdf(mean, std, xmin, xmax)
    thinkplot.Plot(xs, ps, label='model', linewidth=4, color='0.8')
    thinkplot.Cdf(cdf)
```

```python
def MakeNormalPlot(salary):
    """Generates a normal probability plot.

    salary: sequence
    """
    mean, var = thinkstats2.TrimmedMeanVar(salary, p=0.01)
    std = np.sqrt(var)

    xs = [-5, 5]
    xs, ys = thinkstats2.FitLine(xs, mean, std)
    thinkplot.Plot(xs, ys, color='0.8', label='model')

    xs, ys = thinkstats2.NormalProbability(salary)
    thinkplot.Plot(xs, ys, label='salary')
```

```python
def Cov(xs, ys, meanx=None, meany=None):
    xs = np.asarray(xs)
    ys = np.asarray(ys)

    if meanx is None:
        meanx = np.mean(xs)
    if meany is None:
        meany = np.mean(ys)

    cov = np.dot(xs-meanx, ys-meany) / len(xs)
    return cov
```

```python
def Corr(xs, ys):
    xs = np.asarray(xs)
    ys = np.asarray(ys)

    meanx, varx = thinkstats2.MeanVar(xs)
    meany, vary = thinkstats2.MeanVar(ys)

    corr = Cov(xs, ys, meanx, meany) / np.sqrt(varx * vary)
    return corr
```

```python
class DiffMeansPermute(thinkstats2.HypothesisTest):

    def TestStatistic(self, data):
        group1, group2 = data
        test_stat = abs(group1.mean() - group2.mean())
        return test_stat

    def MakeModel(self):
        group1, group2 = self.data
        self.n, self.m = len(group1), len(group2)
        self.pool = np.hstack((group1, group2))

    def RunModel(self):
        np.random.shuffle(self.pool)
        data = self.pool[:self.n], self.pool[self.n:]
        return data
```

```python
data = pitch_df.salary.values, nonPitch_df.salary.values
ht = DiffMeansPermute(data)
pvalue = ht.PValue()
pvalue
```

```python
ht.PlotCdf()
thinkplot.Config(xlabel='test statistic',
                   ylabel='CDF')
```

```python
class DiffMeansOneSided(DiffMeansPermute):

    def TestStatistic(self, data):
        group1, group2 = data
        test_stat = group1.mean() - group2.mean()
        return test_stat
```

```python
ht = DiffMeansOneSided(data)
pvalue = ht.PValue()
pvalue
```
